```python
import pandas as pd

def process_age(df):
    # 1. แปลงเป็น Datetime โดยระบุ dayfirst=True (สำหรับฟอร์แมต dd-mm-yy)
    df['DATE_OF_BIRTH']  = pd.to_datetime(df['DATE_OF_BIRTH'],  dayfirst=True, errors='coerce')
    df['DISBURSAL_DATE'] = pd.to_datetime(df['DISBURSAL_DATE'], dayfirst=True, errors='coerce')

    # 2. แก้ปัญหา Century Bug: ถ้าปีเกิด > ปีที่กู้ (แสดงว่า Pandas ตีความผิดเป็น 20xx) ให้ลบออก 100 ปี
    # เช่น เกิดปี 1965 แต่ Pandas เห็นเป็น 2065 ซึ่งมากกว่าวันกู้ (2018)
    df['DATE_OF_BIRTH'] = df.apply(
        lambda x: x['DATE_OF_BIRTH'] - pd.DateOffset(years=100) 
        if x['DATE_OF_BIRTH'].year > x['DISBURSAL_DATE'].year else x['DATE_OF_BIRTH'], 
        axis=1
    )

    # 3. คำนวณอายุ "ณ วันที่กู้" (แม่นยำระดับวัน)
    df['AGE_AT_DISBURSAL'] = (df['DISBURSAL_DATE'] - df['DATE_OF_BIRTH']).dt.days // 365
    
    # ลบคอลัมน์วันที่เดิมทิ้งเพื่อไม่ให้ Model ติด Error
    df.drop(['DATE_OF_BIRTH'], axis=1, inplace=True)
    return df

# ใช้งานกับทั้ง train และ test
train_df = process_age(train_df)
test_df  = process_age(test_df)

print(train_df[['AGE_AT_DISBURSAL']].head())
```
